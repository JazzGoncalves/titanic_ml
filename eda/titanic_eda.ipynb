```python
train_data['Title'] = train_data['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
rare_titles_train = ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']
train_data['Title'] = train_data['Title'].replace(rare_titles_train, 'Other')
```

```python
test_data['Title'] = test_data['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
rare_titles_test = ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']
test_data['Title'] = test_data['Title'].replace(rare_titles_test, 'Other')
```

Sobrevivência por Título (Title) - Conjunto de Treino

```python
sns.countplot(x='Title', hue='Survived', data=train_data)
plt.title('Sobrevivência por Título - Conjunto de Treino')
plt.xticks(rotation=45)
plt.show()
```

Distribuição por Título - Conjunto de Teste

```python
sns.countplot(x='Title', data=test_data)
plt.title('Distribuição por Título - Conjunto de Teste')
plt.xticks(rotation=45)
plt.show()
```
